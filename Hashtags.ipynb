```python
# import blank week:
#a = np.nan
unique_LHS = list(df_koi['LHS'].unique())
unique_LHS = unique_LHS[1:]

week_number = {}
for i in unique_LHS:
    if i not in week_number:
        week_number[i] = df[(df['RHS'] == '{koi}') & (df['LHS'] == i)]["week"].iloc[0]
#week_number
week_number = {k: v for k, v in sorted(week_number.items(), key=lambda item: item[1])}
week_number
unique_LHS = list(week_number.keys())
unique_LHS
```

```python
import numpy as np

support_list = []
max_len = 328
for tag in unique_LHS:
    occur_week_num = week_number.get(tag)
    nan_list = [np.nan]*(occur_week_num-1)
    curr_list = list(df[(df['RHS'] == '{koi}') & (df['LHS'] == tag) ]["support"])
    remain_len = max_len - len(nan_list) - len(curr_list)
    zero_list = [0] * remain_len
    
    curr_support_list = nan_list + curr_list + zero_list
    
    support_list.append(curr_support_list)

len(support_list[0])
```

```python
import csv
def process_csv(filename):
    example_file = open(filename, encoding="utf-8")
    example_reader = csv.reader(example_file)
    example_data = list(example_reader)
    example_file.close()
    return example_data

#Data Preparation
data = process_csv('final_rules.csv')
title = data[0]
data = data[1:]
```

```python
data[4]
```

```python
support_dict = {}
for tag in unique_LHS:
    if tag not in support_dict:
        support_dict[tag] = {}
    for i in range(len(data)):
        if tag == data[i][4] and data[i][5] == '{koi}':
            week_num = data[i][1]
            if week_num not in support_dict[tag]:
                support_dict[tag][week_num] = float(data[i][6])
support_dict
```

```python
# Change Point Detection
import numpy as np
import ruptures as rpt

new = pd.DataFrame({ key:pd.Series(value) for key, value in support_dict.items() })
new
breakpoint_dict = {}
for key in support_dict:
    if len(list(support_dict[key].values())) == 1:
        continue
    points = np.array(new[key])
    
    points = points[~np.isnan(points)]

#Changepoint detection with the Pelt search method
    model="rbf"
    algo = rpt.Pelt(model=model).fit(points)
    result = algo.predict(pen=10)
    if key not in breakpoint_dict or 'Optimal breakpoints' not in breakpoint_dict[key]:
        breakpoint_dict[key]= {}
        breakpoint_dict[key]['Optimal breakpoints'] = result
    #rpt.display(points, result, figsize=(10, 6))
    #plt.title('Change Point Detection: Pelt Search Method')
    #plt.show() 
breakpoint_dict   


```

```python
CPD_result = pd.DataFrame(breakpoint_dict).T
CPD_result
CPD_result.to_csv('Hashtag_CPD.csv')
```
